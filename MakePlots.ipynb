```julia
# compare TR = [-80,10,0,-10] (cases 1-4) and TR = [-40,40,20,0] (cases 5-8)

# 1. compare avg number of embryos transferred based on starting # viable in culture

simA = [
    mean([simActions1[1:10],simActions2[1:10],simActions3[1:10],simActions4[1:10]]),
    mean([simActions1[11:20],simActions2[11:20],simActions3[11:20],simActions4[11:20]]),
    mean([simActions1[21:30],simActions2[21:30],simActions3[21:30],simActions4[21:30]]),
    mean([simActions1[31:40],simActions2[31:40],simActions3[31:40],simActions4[31:40]])];

simB = [
    mean([simActions5[1:10],simActions6[1:10],simActions7[1:10],simActions8[1:10]]),
    mean([simActions5[11:20],simActions6[11:20],simActions7[11:20],simActions8[11:20]]),
    mean([simActions5[21:30],simActions6[21:30],simActions7[21:30],simActions8[21:30]]),
    mean([simActions5[31:40],simActions6[31:40],simActions7[31:40],simActions8[31:40]])];

a=Axis([
    Plots.Linear([0.15,1.15,2.15,3.15],simA,legendentry=L"TR=[-80,80,0,0]",style="ybar,fill=blue,fill opacity=0.3"),
    Plots.Linear([-.15,.85,1.85,2.85],simB,legendentry=L"TR=[-40,40,20,0]",style="ybar,fill=red,fill opacity=0.3")
    ], xlabel="number of viable embryos", ylabel="average number transferred", 
title="Transfer Policy",legendPos="north west", ymax=2.6)

#save("transfer_policy.pdf", a)
```

```julia
# 2. Compare percentage of transfers resulting in singleton pregnancy, 
#    twin (or more pregnancy), no pregnancy, and percentage of time all were discarded

simActionsA = [simActions1,simActions2,simActions3,simActions4];
simActionsB = [simActions5,simActions6,simActions7,simActions8];
simRewardsA = [simRewards1,simRewards2,simRewards3,simRewards4];
simRewardsB = [simRewards5,simRewards6,simRewards7,simRewards8];

# D,V (-45,-40); D,NV (-5,0); T,NP (-85,-80); T,1 (5,10); T,2+ (-15,0)
pA = [
    sum((simActionsA .== 0) & (simRewardsA .< -20)),
    sum((simActionsA .== 0) & (simRewardsA .> -20) & (simRewardsA .< 1)),
    sum((simActionsA .> 0) & (simRewardsA .< -20)),
    sum((simActionsA .> 0) & (simRewardsA .> 1)),
    sum((simActionsA .> 0) & (simRewardsA .> -20) & (simRewardsA .< 1))
    ] / 160.0;

# D,V (-45,-40); D,NV (-5,0); T,NP (-45,-40); T,1 (35,40); T,2+ (-5,20)
pB = [
    sum((simActionsB .== 0) & (simRewardsB .< -20)),
    sum((simActionsB .== 0) & (simRewardsB .> -20) & (simRewardsB .< 1)),
    sum((simActionsB .> 0) & (simRewardsB .< -20)),
    sum((simActionsB .> 0) & (simRewardsB .> 30)),
    sum((simActionsB .> 0) & (simRewardsB .> -20) & (simRewardsB .< 30))
    ] / 160.0;


a=Axis([
    Plots.Linear([0.15,1.15,2.15,3.15,4.15],pA,legendentry=L"TR=[-80,10,0,-10]",style="ybar,fill=blue,fill opacity=0.3"),
    Plots.Linear([-.15,.85,1.85,2.85,3.85],pB,legendentry=L"TR=[-40,40,20,0]",style="ybar,fill=red,fill opacity=0.3")
    ], ylabel="percent of simulations", title="Transfer Outcomes",legendPos="north west", ymax=.6) 

#save("reward_policy.pdf", a)
```
